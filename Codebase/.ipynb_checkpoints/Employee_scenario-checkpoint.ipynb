```python
class Employee:
    def __init__(self, empId, empName, UANNo, AadharNo, PanNo, Course1, Course2, Rating):
        self.empId = empId
        self.empName = empName
        self.UANNo = UANNo
        self. AadharNo = AadharNo
        self.PanNo = PanNo
        self.Course1 = Course1
        self.Course2 = Course2
        self.Rating = Rating
```

```python
class LowRatingException (Exception):
    pass
```

```python
def check_eligibility (employee):
    if employee.Rating > 90:
        if (employee.Course1.lower() == 'SQL' and employee.Course2.lower() == 'Python') or (employee.Course1.lower() == 'Python' and employee.Course2.lower() == 'SQL'):
            return True
        else:
            raise LowRatingException (f"{employee.empId} - {employee.empName}'s Rating is {math.ceil(100 - (employee.Rating/90) * 100)}% less than the Eligibility criteria")
    else:
            raise LowRatingException (f"{employee.empId} - {employee.empName}'s Rating is {math.ceil(100 - (employee.Rating/90) * 100)}% less than the Eligibility criteria")
```

```python
def allocate_project (employees):
    eligibility_list = []
    for emp in employees:
        try:
            if check_eligibility(emp):
                eligibility_list.append(emp)
        except LowRatingException as e:
            print(e)
    return eligibility_list
```

```python
import math
if __name__ == "__main__":
    
    end = False
    employees = [
            Employee(1, "John", 12345, "123456789", "ABCDE1234F","SQL", "Python", 25),
            Employee(2, "barn", 12345, "123456789", "ABCDE1234F","SQL", "Python", 95),
            Employee(3, "Tarn", 12345, "123456789", "ABCDE1234F","SQL", "Python", 95),
    ]
    
    while not end :
        
        print("\n1. Add Employee Details\n2. Generate Report\n3. Exit\n   Enter ur option: ", end = " ")
        option = int(input())
        
        if option == 1:
            
            print("Emp ID   Emp Name   UAN No   Aadhar No   PAN No   Course 1   Course 2   Rating")
            id, name, uan, aadhar, pan, c1, c2, rating = input().split(" ")
            
            employees.append(Employee(id, name, uan, int(aadhar), pan, c1, c2, float(rating)))
            
            print("**** SUCCESSFULLY ADDED ****")
            
        elif option == 2:
            
            print("\n*************** R E P O R T *******************\n")
            Shortlisted = allocate_project(employees)
            
            print("\n*********** SHORT-LISTED EMPLOYEES ************\n")
            
            for emp in Shortlisted:
                print(f"Employee ID - {emp.empId}\nEmployee Name - {emp.empName}\nRating - {emp.Rating}\n")
                
            print("\n***********************************************\n")
            
        else:
            
            end = True
            print("**** EXITING ****")

```
