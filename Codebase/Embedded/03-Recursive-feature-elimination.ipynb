# Recursive feature elimination - feature importance

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

```python
# load dataset

df = pd.read_csv("C:/Users/H A R I H A R A N/Desktop/sem 8/Heart Disease/heart.csv")
x = df.drop('target',axis = 1)
y = df['target']

# Separate data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=135)
```

```python
X_train.head()
```

### Select features recursively

```python
a = []
b = []
for i in range(1,14):
    sel_ = RFE(
        RandomForestClassifier(n_estimators=10, random_state=135),
        n_features_to_select=i,
        step=2)
    sel_.fit(X_train, y_train)
    selected_feat = X_train.columns[(sel_.get_support())]
    selected_feat = list(selected_feat)
    selected_feat.append('target')
    a.append(selected_feat)
    
    new = df[selected_feat]
    x=new.drop('target',axis=1)    
    y=new['target']
    x_train1,x_test1,y_train1,y_test1 = train_test_split(x, y, test_size=0.20, random_state = 135)
    rf_Model = RandomForestClassifier() 
    model = rf_Model.fit(x_train1,y_train1)
    d = model.score(x_test1,y_test1)*100
    b.append(d)
```

```python
sel_ = RFE(
    RandomForestClassifier(n_estimators=10, random_state=135),
    n_features_to_select=12,
    step=2,
)

sel_.fit(X_train, y_train)
```
