```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acc
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
df = pd.read_csv("C:/Users/H A R I H A R A N/Desktop/sem 8/Heart Disease/heart.csv")
df
```

```python
Y = df['target']
X = df.drop(['target'], axis = 1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
test = SelectKBest(score_func=chi2, k=13)
fit = test.fit(X, Y)
np.set_printoptions(precision=3)
print(fit.scores_)
chi = np.argsort(fit.scores_)[::-1]
chi
```

```python
l = list(df.columns)
l.pop()
l
```

```python
chisq_features = []
for i in chi:
    chisq_features.append(l[i])
chisq_features
```

```python
gscv = pd.DataFrame(columns = np.arange(1,14))
```

```python
for i in range(15):
    p = []
    for i in range(1,14):
        X = df[chisq_features[:i]]
        Y = df[['target']]
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state= 135)  
        rf_Model = RandomForestClassifier() 
        model = rf_Model.fit(X_train,y_train)
        o = model.score(X_test,y_test)*100
        p.append(o)
    print(p)
    gscv.loc[len(gscv)] = p
```

```python
gscv
```

```python
avg = gscv.mean()
display(avg)
```
