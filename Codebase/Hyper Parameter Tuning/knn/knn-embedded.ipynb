```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/H A R I H A R A N/Desktop/sem 8/Heart Disease/heart.csv")
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
new = df[['sex','cp','fbs','oldpeak','slope','ca','thal','target']]


Y = new['target']
X = new.drop(['target'], axis = 1)

x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.20, random_state = 65)
```

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
rf_params = {
    'n_neighbors': range(1,20),
}
clf = KNeighborsClassifier()
grid = GridSearchCV(clf, rf_params, cv=3, scoring='accuracy')
grid.fit(x_train, y_train)
print(grid.best_params_)
print("Accuracy:"+ str(grid.best_score_))
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.20, random_state = 65)
rf_Model = KNeighborsClassifier(n_neighbors=6) 
model = rf_Model.fit(x_train,y_train)
print (model.score(x_test,y_test)*100, )
```

```python
#KNN
from scipy.stats import randint as sp_randint
from sklearn.model_selection import RandomizedSearchCV
rf_params = {
    'n_neighbors': range(1,20),
}
n_iter_search=10
clf = KNeighborsClassifier()
Random = RandomizedSearchCV(clf, param_distributions=rf_params,n_iter=n_iter_search,cv=3,scoring='accuracy')
Random.fit(x_train, y_train)
print(Random.best_params_)
print("Accuracy:"+ str(Random.best_score_))
```

```python
rf_Model = KNeighborsClassifier(n_neighbors=7) 
model = rf_Model.fit(x_train,y_train)
print (model.score(x_test,y_test)*100)
```
