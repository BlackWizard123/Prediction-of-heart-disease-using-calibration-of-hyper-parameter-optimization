```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/H A R I H A R A N/Desktop/sem 8/Heart Disease/heart.csv")
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
new = df[['cp',
 'thal',
 'ca',
 'slope',
 'exang',
 'sex','target']]

Y = new['target']
X = new.drop(['target'], axis = 1)

x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.20, random_state = 135)
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
rf_Model = XGBClassifier() 
model = rf_Model.fit(x_train,y_train)
xgb_params = [{'learning_rate': [0.1],
 'n_estimators': [100,200,300,400,500,600,700,800,900],
 'max_depth': [2,3,4],
 'min_child_weight': [7,8,9],
 'gamma': [1],
 'subsample': [0.8]}]
clf = GridSearchCV(model, xgb_params, cv = 10, scoring='accuracy')
best_clf = clf.fit(x_train,y_train)
best_clf.best_estimator_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state = 135)
xgb_Model = XGBClassifier( learning_rate =0.1,n_estimators=400, max_depth=2, min_child_weight=7, gamma=1, subsample=0.8)
model = xgb_Model.fit(X_train,y_train)
model.score(X_test,y_test)*100
```
